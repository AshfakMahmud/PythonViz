```python
#full heatmap 

df=pd.read_csv('heatmap_16_17.csv')
plt.figure(figsize=(16, 6))

heatmap = sns.heatmap(df.corr(),  annot=True, cmap='BrBG_r') # add linewidths=1 if you want space in between
```

```python
#half heatmap

df=pd.read_csv('heatmap_16_17.csv')
plt.figure(figsize=(16, 6))



np.triu(np.ones_like(df.corr()))
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

heatmap = sns.heatmap(df.corr(),  annot=True, mask=mask, cmap='BrBG_r') # add linewidths=1 if you want space in between
```
